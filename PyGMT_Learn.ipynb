```python
import pygmt
```

```python
import matplotlib.pyplot as plt
```

```python
minlon, maxlon = 60, 95
minlat, maxlat = 0, 25
```

```python
topo_data = '@earth_relief_30s'
```

```python
fig = pygmt.Figure()
```

```python
# make color pallets
pygmt.makecpt(
    cmap='topo',
    series='-8000/8000/1000',
    continuous=True
)
```

```python
fig.grdimage(
    grid=topo_data,
    region=[minlon, maxlon, minlat, maxlat],
    projection='M4i',
    shading=True,
    frame=True
    )
```

```python
fig.show()
```
